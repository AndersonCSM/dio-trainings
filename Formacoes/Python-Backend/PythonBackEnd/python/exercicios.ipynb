```python
import csv
from pathlib import Path

ROOT_PATH = Path().resolve()

try:
    with open(f"{ROOT_PATH}/exemplo.csv", "w", encoding="utf-8", newline="") as file:
        wrt = csv.writer(file)
        wrt.writerow(["id", "nome"])
        wrt.writerow(["1", "Maria"])
        wrt.writerow(["2", "Joáo"])
except IOError as exc:
    print(f"Erro ao criar o arquivo: \n{exc}")
```
